```python
import statsmodels.api as sm
# Take raw 4 and 19
Y_raw = df.iloc[3][1:]
X_raw = df.iloc[18][1:]

# replace ',' to ''
Y = Y_raw.str.replace(',', '').astype(float)
X = X_raw.str.replace(',', '').astype(float)

# delet NaN
mask = (~X.isna()) & (~Y.isna())
X = X[mask]
Y = Y[mask]

#  add constant
X = sm.add_constant(X)

# model
model = sm.OLS(Y, X).fit()

# Output regression coefficient and R^2
coefficients = model.params
intercept = coefficients.loc['const']      
slope = coefficients.loc[X.columns[1]]      

r_squared = model.rsquared

print(f"regression equation Y = {intercept:.4f} + {slope:.4f} * X")
print(f"R² = {r_squared:.4f}")
```
